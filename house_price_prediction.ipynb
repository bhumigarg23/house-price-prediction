```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
num_cols = [col for col in num_cols if df[col].isnull().sum() > 0]

print(num_cols)
```

```python
num_cols = ["LotFrontage", "GarageYrBlt", "MasVnrArea"]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = [
    "GarageType", "GarageFinish", "GarageQual", "GarageCond",
    "BsmtExposure", "BsmtFinType2", "BsmtQual",
    "BsmtCond", "BsmtFinType1", "Electrical"
]

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
Q1 = df["SalePrice"].quantile(0.25)
Q3 = df["SalePrice"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["SalePrice"] >= lower) & (df["SalePrice"] <= upper)]
```
